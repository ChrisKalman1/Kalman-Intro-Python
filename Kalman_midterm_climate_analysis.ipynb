## Part 0: Setup and Data Loading (IMPORTANT - Do This First!)

````python
import csv
import os

# Define the directory path
DATA_PATH = '/Users/chriskalman/Downloads/python_spring/Python Midterm'

# Load the CSV file
def load_climate_data(filename):
    """
    Loads the CSV file and returns a list of dictionaries.
    Each dictionary represents one row of data.
    """
    # Construct the full file path
    full_path = os.path.join(DATA_PATH, filename)
    
    data = []
    try:
        with open(full_path, 'r', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            for row in reader:
                data.append(row)
        print(f"Successfully loaded {len(data)} rows of data from {full_path}!")
        return data
    except FileNotFoundError:
        print(f"Error: Could not find {full_path}")
        print(f"Make sure the CSV file is in the folder: {DATA_PATH}")
        return []
    except Exception as e:
        print(f"Error loading file: {e}")
        return []

# Load the data
climate_data = load_climate_data('owid-co2-data.csv')

# Show the first row to see what data we have
if climate_data:
    print("\nSample row (first entry):")
    print(climate_data[0])
    print("\nAvailable columns:")
    print(list(climate_data[0].keys()))
````

### Problem 1.2: Safe Data Conversion (7 points)

Many values in the CSV are stored as strings, but we need them as numbers. Some values are missing (empty strings).

Write a function called `safe_float_convert(value, default=0.0)` that:
- Takes a string value and an optional default parameter
- Tries to convert the value to a float
- If conversion fails (empty string, invalid data, etc.), returns the default value
- Uses try-except for error handling

**Example usage:**
```python
print(safe_float_convert("123.45"))  # Should print 123.45
print(safe_float_convert(""))        # Should print 0.0
print(safe_float_convert("abc"))     # Should print 0.0
print(safe_float_convert("", -1))    # Should print -1
```

````python
# Your code here:
def safe_float_convert(value, default=0.0):
    try:
        return float(value)
    except (ValueError, TypeError):
        return default
print(safe_float_convert("123.45"))
print(safe_float_convert(""))
print(safe_float_convert("abc"))
print(safe_float_convert("", -1))
````

- Returns a dictionary with keys: 'year', 'value', and 'country'


````python
# Your code here:
def get_latest_year_data(country_rows, indicator):
    sorted_rows = sorted(country_rows, key=lambda row: row['year'], reverse = True)
    for row in sorted_rows:
        if row[indicator] != "":
            value = safe_float_convert(row[indicator])
            if value != 0.0:
                return {
                    'year': row['year'],
                    'value': value,
                    'country': row['country']
                }
china_data = get_country_data(climate_data, 'China')
latest_co2 = get_latest_year_data(china_data, 'co2')
if latest_co2:
    print(f"{latest_co2['country']} emitted {latest_co2['value']} million tonnes of CO2 in {latest_co2['year']}")
````

### Problem 2.1: Create Multi-Country Database (10 points)

Write a function called `build_country_database(data, countries, year)` that:

```python

        'renewables_share_energy': 12.5,
        'fossil_share_energy': 80.2
    },
    'China': {
        ...
    }
}
```
- Uses safe_float_convert to handle missing data
- If a country doesn't have data for that year, skip it (don't include in dictionary)
- The indicators to include are: 'co2', 'co2_per_capita', 'population', 'renewables_share_energy', 'fossil_share_energy'

**Test your function with a list of at least 10 countries for the year 2020**

````python
# Your code here:
test_countries = ['United States', 'China', 'Germany', 'India', 'Brazil', 
                  'United Kingdom', 'France', 'Canada', 'Japan', 'Australia']

def build_country_database(data, countries, year):
    result = {}
    for country in countries:
        for row in data:
            if row['country'] == country and row ['year'] == str(year):
                country_data = {
                    'co2': safe_float_convert(row['co2']),
                    'co2_per_capita': safe_float_convert(row['co2_per_capita']),
                    'population': safe_float_convert(row['population']),
                    'oil_co2': safe_float_convert(row['oil_co2']),
                    'energy_per_capita': safe_float_convert(row['energy_per_capita'])
                }
                if country_data['co2'] !=0.0 or country_data['population'] !=0.0:
                    result[country] = country_data
                break
    return result

db_2020 = build_country_database(climate_data, test_countries, 2020)

for country, data in db_2020.items():
    print(f"{country}: CO2={data['co2']}, Population={data['population']}")
````

Create a function called `group_by_region(country_database, region_definitions)` that:

**Important:** Use try-except to handle countries that might not be in your database. Calculate averages only from countries that have valid data.

````python
# Your code here:
regions = {
    'North America': ['United States', 'Canada', 'Mexico'],
    'Europe': ['Germany', 'France', 'United Kingdom', 'Italy', 'Spain'],
    'Asia': ['China', 'India', 'Japan', 'South Korea', 'Indonesia'],
    'South America': ['Brazil', 'Argentina', 'Colombia', 'Chile'],
    'Africa': ['South Africa', 'Nigeria', 'Egypt', 'Kenya'],
    'Oceania': ['Australia', 'New Zealand']
}

test_countries = ['United States', 'China', 'Germany', 'India', 'Brazil', 
                  'United Kingdom', 'France', 'Canada', 'Japan', 'Australia',
                  'Mexico', 'Italy', 'Spain', 'South Korea', 'Indonesia',
                  'Argentina', 'Colombia', 'Chile', 'South Africa', 'Nigeria',
                  'Egypt', 'Kenya', 'New Zealand']

db_2020 = build_country_database(climate_data, test_countries, 2020)

def group_by_region(country_database, region_definitions):
    result = {}

    for region_name, country_list in region_definitions.items():
        total_co2 = 0
        total_population = 0
        total_co2_per_capita = 0
        countries_with_data = 0

        for country in country_list:
            try:
                if country in country_database:
                    total_co2 += country_database[country]['co2']
                    total_population += country_database[country]['population']
                    total_co2_per_capita += country_database[country]['co2_per_capita']
                    countries_with_data += 1
            except:
                continue
        if countries_with_data > 0:
            avg_co2_per_capita = total_co2_per_capita / countries_with_data
        else:
            avg_co2_per_capita = 0

        result[region_name] = {
            'total_co2': total_co2,
            'avg_co2_per_capita': avg_co2_per_capita,
            'total_population': total_population,
            'countries': country_list
        }
    return result

regional_data = group_by_region(db_2020, regions)

for region, data in regional_data.items():
    print(f"\n{region}:")
    print(f"  Total CO2: {data['total_co2']:.2f} million tonnes")
    print(f"  Average per capita: {data['avg_co2_per_capita']:.2f} tonnes")
    print(f"  Total population: {data['total_population']:,.0f}")
    print(f"  Countries: {', '.join(data['countries'])}")

````

**Renewables_share is not a real variable so I am replacing it with coal_co2_per_capita**

````python
### finding averages to make classifications more accurate

values = [
    safe_float_convert(row['coal_co2_per_capita'])
    for row in climate_data
    if row['year'] == '2024'
]

avg = sum(values) / len(values) if values else 0
print("Average coal CO2 per capita in 2024:", avg)

values_co = [
    safe_float_convert(row['co2_per_capita'])
    for row in climate_data
    if row['year'] == '2024'
]

avg = sum(values_co) / len(values_co) if values else 0
print("Average  CO2 per capita in 2024:", avg)
````
