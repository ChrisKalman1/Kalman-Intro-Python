````python
# TODO: write your class here
# I am going to create a class that will store information relevant to restoration practices for a tree

class tree:
    def __init__(self, ID_number, species, height, dbh, alive=True):
        self.ID_number = ID_number
        self.species = species
        self.height = height
        self.alive = alive
        self.dbh = dbh

    def describe(self):
        status = "alive" if self.alive else "dead"
        print(f"This is an {self.species} that is {self.height} feet tall, with a diameter breast height (dbh) of {self.dbh} inches, and is currently {status}.")

    def toggle(self):
        self.alive = not self.alive
        status = "alive" if self.alive else "dead"
        print(f" The tree is now {status}")

    def update(self, new_entry):
        if not isinstance(new_entry,int):
            raise ValueError("Height must be an integer")
            
        old_height = self.height
        self.height = new_entry
        print(f"Tree height is now {self.height}, instead of {old_height}.")

    def compare(self, other):
        if self.height > other.height:
            print(f"This {self.species} tree ({self.height}ft) is taller than the {other.species} tree ({other.height}ft).")
        elif self.height < other.height:
            print(f"This {self.species} tree ({self.height}ft) is shorter than the {other.species} tree ({other.height}ft).")
        else:
            print(f"These trees are both {self.height} ft. tall.")

    def volume(self):
        volume = (self.height * self.dbh)
        print(f"This tree has a volume of {volume} square feet.")
         
    
````

> ### 📘 New Concept — `getattr(obj, name)`
>
> So far you have accessed attributes directly: `obj.speed`, `obj.name`, etc.
> That works fine when you know the attribute name at the time you write the code.
> But what if the name is stored in a variable — chosen at runtime by the user?
>
> `getattr()` solves this. It takes an object and a **string** containing the
> attribute name, and returns that attribute's value:
>
> ```python
> # These two lines do exactly the same thing:
> value = obj.speed          # attribute name is hardcoded
> value = getattr(obj, 'speed')  # attribute name is a string
> ```
>
> Because the name is just a string, you can pass it in as a variable:
>
> ```python
> attribute = input('Which attribute? ')   # e.g. user types 'speed'
> value = getattr(obj, attribute)          # looks up obj.speed at runtime
> ```
>
> If the attribute does not exist, `getattr()` raises an `AttributeError` —
> which you can catch with a `try-except`, as you will do in the question below.
>
> You can also supply a **fallback default** as a third argument, so no error is raised:
> ```python
> value = getattr(obj, attribute, None)  # returns None if attribute doesn't exist
> ```

## Question 4 — Searching & Filtering ⭐⭐⭐⭐

A registry is only useful if you can find things in it quickly. In this question you will write two search tools.

---

### Tool 1 — `search_by_attribute(registry, attribute, value)`

- Loops through every object in the registry.
- Collects all objects where `getattr(obj, attribute)` equals `value` — use `getattr()`, do not hardcode the attribute name.
- Returns a **list** of matching objects.
- If **no matches** are found → print an informative message and return an empty list.
- If the **attribute does not exist** on the objects → catch the `AttributeError` and print a helpful message.

---

### Tool 2 — `filter_by_range(registry, attribute, min_val, max_val)`

- Works only on **numeric** attributes.
- Returns a list of objects where the attribute's value falls within `[min_val, max_val]` **inclusive**.
- Raises a `TypeError` with a clear message if the attribute's value is not a number.
- If the result list is **empty** → print a message saying no objects fell within that range.

---

### Demonstrate both tools:
- Call `search_by_attribute` with at least **two different attributes**.
- Call `filter_by_range` including one test where **the range contains no results**.
- For each result list, print the results using your `describe` method.

````python
# TODO: write search_by_attribute here
def search_by_attribute(registry,attribute,value):
    matches = []

    for key,obj in registry.items():
        try:
            obj_value = getattr(obj,attribute)
            if obj_value == value:
                matches.append(obj)
        except AttributeError:
            print(f"Error: Objects in registry don't have attribute '{attribute}'")
            return[]

    if len(matches)==0:
        print(f"No trees found with {attribute} = {value}")
        return[]


    return matches
````

````python
# TODO: write filter_by_range here
def filter_by_range(registry, attribute, min_val, max_val):
    results = []

    for key, obj in registry.items():
        try:
            obj_value = getattr(obj, attribute)

            if not isinstance(obj_value, (int, float)):
                raise TypeError(f"Attribute '{attribute}' is not numeric")

            if obj_value >= min_val and obj_value <= max_val:
                results.append(obj)

        except AttributeError:
            print(f"Error: Object doesn't have attribute '{attribute}'.")
            return []

        except TypeError as e:
            print(f"Error: {e}")
            return []

    if len(results) == 0:
        print(f"No trees found with {attribute} between {min_val} and {max_val}")
        return []

    return results
````

> ### 📘 New Concept — Subclasses & Inheritance
>
> A **subclass** is a class that is built on top of an existing one.
> The existing class is called the **parent** (or base) class; the new one is the **child** (or subclass).
> The child automatically inherits everything the parent has — all its attributes and methods —
> without you having to rewrite any of it.
>
> You create a subclass by passing the parent class as an argument:
>
> ```python
> class Animal:                      # parent class
>     def __init__(self, name, age):
>         self.name = name
>         self.age  = age
>
>     def describe(self):
>         print(f'{self.name}, age {self.age}')
>
> class Dog(Animal):                 # Dog is a subclass of Animal
>     def __init__(self, name, age, breed):
>         super().__init__(name, age) # run the parent's __init__ first
>         self.breed = breed          # then add what's new
>
>     def fetch(self):               # a method only Dogs have
>         print(f'{self.name} fetches the ball!')
> ```
>
> **`super()`** is how the child talks to its parent. Calling `super().__init__()`
> runs the parent's setup code so you don't have to copy it into the child.
>
> The child can also **override** a parent method by defining one with the same name.
> Python will always use the child's version when the method is called on a child object:
>
> ```python
>     def describe(self):                      # overrides Animal.describe
>         super().describe()                   # optionally call the parent's version first
>         print(f'Breed: {self.breed}')        # then add the extra detail
> ```
>
> This means a `Dog` object has access to everything `Animal` defines,
> plus anything extra added in `Dog` itself. A plain `Animal`, however,
> knows nothing about `Dog` — inheritance only flows **downward**.
>
> When you loop over a collection containing both `Animal` and `Dog` objects
> and call `describe()` on each one, Python automatically calls the right version
> for each object. This behaviour is called **polymorphism**, and it is what
> Question 5 Part C asks you to observe.

````python
# TODO: write your subclass here
class leaf(tree):
    def __init__(self, ID_number, species, height, dbh, alive=True, color=None, lobes=0):
        super().__init__(ID_number, species, height, dbh, alive)
        self.color = color
        self.lobes = lobes
    
    def describe(self):
        super().describe()
        print(f'Leaf Color: {self.color}. Leaf Lobes: {self.lobes}')
    
    def compare(self, other):
        if self.lobes > other.lobes:
            print(f"This leaf ({self.lobes} lobes) has more lobes than the other leaf ({other.lobes} lobes).")
        elif self.lobes < other.lobes:
            print(f"This leaf ({self.lobes} lobes) has fewer lobes than the other leaf ({other.lobes} lobes).")
        else:
            print(f"Both leaves have the same amount of lobes ({self.lobes}).")
    
    def analyze_leaf(self):
        if self.color == "green":
            status = "healthy"
        elif self.color == "brown":
            status = "dead"
        else:
            status = "unknown"
        print(f"This {self.color} leaf with {self.lobes} lobes appears to be {status}.")
````

````python
# TODO: write each menu handler function here

def get_valid_string(prompt):
    """Get a non-empty string from user."""
    while True:
        try:
            value = input(prompt).strip()
            if not value:
                print("Error: Input cannot be empty. Please try again.")
                continue
            return value
        except Exception as e:
            print(f"Error: {e}. Please try again.")

def get_valid_int(prompt):
    """Get a valid integer from user."""
    while True:
        try:
            value = int(input(prompt))
            return value
        except ValueError:
            print("Error: Please enter a valid integer.")

def get_valid_boolean(prompt):
    """Get a valid boolean (yes/no or true/false) from user."""
    while True:
        try:
            value = input(prompt).lower().strip()
            if value in ["yes", "true", "y", "t"]:
                return True
            elif value in ["no", "false", "n", "f"]:
                return False
            else:
                print("Error: Please enter 'yes' or 'no' (or 'true'/'false').")
        except Exception as e:
            print(f"Error: {e}")

def handle_add(registry):
    """Prompt user to add a new tree to the registry."""
    print("\n=== Add a New Entry ===")
    
    try:
        # Get tree attributes
        ID_number = get_valid_int("Enter ID number: ")
        species = get_valid_string("Enter species name: ")
        height = get_valid_int("Enter height (feet): ")
        dbh = get_valid_int("Enter DBH (inches): ")
        alive = get_valid_boolean("Is the tree alive? (yes/no): ")
        
        # Ask if it's a leaf tree
        is_leaf = get_valid_boolean("Is this a leaf tree? (yes/no): ")
        
        if is_leaf:
            color = get_valid_string("Enter leaf color: ")
            lobes = get_valid_int("Enter number of lobes: ")
            new_object = leaf(ID_number, species, height, dbh, alive, color, lobes)
        else:
            new_object = tree(ID_number, species, height, dbh, alive)
        
        add_to_registry(registry, new_object)
    
    except Exception as e:
        print(f"Error adding entry: {e}")

def handle_view(registry):
    """Display all entries in the registry."""
    print("\n=== All Entries in Registry ===")
    
    if len(registry) == 0:
        print("The registry is empty. Add some trees!")
        return
    
    print(f"Total entries: {len(registry)}\n")
    
    for i, (key, obj) in enumerate(registry.items(), 1):
        print(f"{i}. ID {key}:")
        obj.describe()
        print()

def handle_search(registry):
    """Search for entries by attribute."""
    print("\n=== Search by Attribute ===")
    print("Available attributes: species, height, dbh, alive, color, lobes")
    
    try:
        attribute = get_valid_string("Enter attribute to search: ")
        value_str = get_valid_string("Enter value to search for: ")
        
        if attribute in ["height", "dbh", "lobes"]:
            value = int(value_str)
        elif attribute == "alive":
            value = get_valid_boolean("Enter value (yes/no): ")
        else:
            value = value_str
        
        results = search_by_attribute(registry, attribute, value)
        
        if len(results) > 0:
            print(f"\nFound {len(results)} result(s):\n")
            for i, obj in enumerate(results, 1):
                print(f"{i}.")
                obj.describe()
                print()
    
    except Exception as e:
        print(f"Error searching: {e}")

def handle_filter(registry):
    """Filter entries by numeric range."""
    print("\n=== Filter by Numeric Range ===")
    print("Available numeric attributes: height, dbh, lobes")
    
    try:
        attribute = get_valid_string("Enter numeric attribute: ")
        min_val = get_valid_int("Enter minimum value: ")
        max_val = get_valid_int("Enter maximum value: ")
        
        results = filter_by_range(registry, attribute, min_val, max_val)
        
        if len(results) > 0:
            print(f"\nFound {len(results)} result(s):\n")
            for i, obj in enumerate(results, 1):
                print(f"{i}.")
                obj.describe()
                print()
    
    except Exception as e:
        print(f"Error filtering: {e}")

def handle_edit(registry):
    """Edit an existing entry."""
    print("\n=== Edit an Entry ===")
    
    try:
        key = get_valid_int("Enter ID number to edit: ")
        
        if key not in registry:
            print(f"Error: No entry with ID {key} found.")
            return
        
        species = get_valid_string("Enter new species name: ")
        height = get_valid_int("Enter new height (feet): ")
        dbh = get_valid_int("Enter new DBH (inches): ")
        alive = get_valid_boolean("Is the tree alive? (yes/no): ")
        
        if isinstance(registry[key], leaf):
            color = get_valid_string("Enter new leaf color: ")
            lobes = get_valid_int("Enter new number of lobes: ")
            updated_object = leaf(key, species, height, dbh, alive, color, lobes)
        else:
            updated_object = tree(key, species, height, dbh, alive)
        
        edit_registry(registry, key, updated_object)
    
    except Exception as e:
        print(f"Error editing entry: {e}")

def handle_remove(registry):
    """Remove an entry from the registry."""
    print("\n=== Remove an Entry ===")
    
    try:
        key = get_valid_int("Enter ID number to remove: ")
        
        if key not in registry:
            print(f"Error: No entry with ID {key} found.")
            return
        
        print("\nEntry to be deleted:")
        registry[key].describe()
        
        confirm = get_valid_boolean("\nAre you sure you want to delete this? (yes/no): ")
        
        if confirm:
            del registry[key]
            print(f"Successfully deleted entry with ID {key}.")
        else:
            print("Deletion cancelled.")
    
    except Exception as e:
        print(f"Error removing entry: {e}")

def handle_stats(registry):
    """Display statistics about the registry."""
    print("\n=== Statistics ===")
    
    if len(registry) == 0:
        print("The registry is empty. Add some trees to see statistics!")
        return
    
    # Total entries
    print(f"Total entries: {len(registry)}")
    
    # Highest and lowest height
    heights = [obj.height for obj in registry.values()]
    dbhs = [obj.dbh for obj in registry.values()]
    
    max_height = max(heights)
    min_height = min(heights)
    avg_height = round(sum(heights) / len(heights), 2)
    
    print(f"\nHeight Statistics:")
    print(f"  Tallest tree: {max_height} feet")
    print(f"  Shortest tree: {min_height} feet")
    print(f"  Average height: {avg_height} feet")
    
    # DBH statistics
    max_dbh = max(dbhs)
    min_dbh = min(dbhs)
    avg_dbh = round(sum(dbhs) / len(dbhs), 2)
    
    print(f"\nDBH Statistics:")
    print(f"  Largest DBH: {max_dbh} inches")
    print(f"  Smallest DBH: {min_dbh} inches")
    print(f"  Average DBH: {avg_dbh} inches")
    
    # Alive vs Dead
    alive_count = sum(1 for obj in registry.values() if obj.alive)
    dead_count = len(registry) - alive_count
    
    print(f"\nTree Health:")
    print(f"  Alive: {alive_count}")
    print(f"  Dead: {dead_count}")
````

````python
def run_manager():
    """
    Entry point for the interactive registry manager.
    Initialises an empty registry and runs the menu loop
    until the user chooses to quit.
    """
    registry = {}

    print("Tree Registry")
    
    while True:
        print("\n" + "="*40)
        print("MAIN MENU")
        print("="*40)
        print("[1] Add a new entry")
        print("[2] View all entries")
        print("[3] Search by attribute")
        print("[4] Filter by numeric range")
        print("[5] Edit an entry")
        print("[6] Remove an entry")
        print("[7] Show statistics")
        print("[0] Quit")
        print("="*40)
        
        try:
            choice = input("Enter your choice (0-7): ").strip()
            
            if choice == "1":
                handle_add(registry)
            elif choice == "2":
                handle_view(registry)
            elif choice == "3":
                handle_search(registry)
            elif choice == "4":
                handle_filter(registry)
            elif choice == "5":
                handle_edit(registry)
            elif choice == "6":
                handle_remove(registry)
            elif choice == "7":
                handle_stats(registry)
            elif choice == "0":
                print("Closing")
                break
            else:
                print("Error: Invalid choice. Please enter 0-7.")
        
        except KeyboardInterrupt:
            print("\n\nProgram interrupted. Goodbye!")
            break
        except Exception as e:
            print(f"Unexpected error: {e}. Please try again.")

run_manager()
````
